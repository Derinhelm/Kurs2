```python
class Node:
    def __init__(self, id1, dom1, feat1, text1):
        self.id = id1
        self.dom = dom1
        self.feat = feat1
        self.text = text1
```

```python
def parseXML(nameFile):
    doc = xml.dom.minidom.parse(nameFile)
    # а зачем храним id(на всякий случай), вообще-то можно и не хранить, id = index + 1
    arrayParseSentences = []
    parent = doc.getElementsByTagName('S')
    for item in parent:
        newSentence = []
        for child in item.getElementsByTagName('W'):
            newNode = Node(child.getAttribute('ID'), child.getAttribute('DOM'), child.getAttribute('FEAT'), child.getAttribute('LEMMA'))
            newSentence.append(newNode)
        arrayParseSentences.append(newSentence)
    return arrayParseSentences
```

```python
class Inserter:
    def __init__(self, mf, mt, df, dt):
        self.mainFeat = mf
        self.mainText = mt
        self.depFeat = df
        self.depText = dt
        self.count = 1
    def __eq__(self, other):
        return self.mainFeat == other.mainFeat and self.depFeat == other.depFeat and \
            self.mainText == other.mainText and self.depText == other.depText
```

```python
class OnlyTwoMorf:
    def __init__(self, mf, df, p):
        self.mainFeat = mf
        self.depFeat = df
        self.count = 1
    def __eq__(self, other):
        if self.mainFeat == other.mainFeat and self.depFeat == other.depFeat and self.prep == other.prep:
            return True
        return False
```

```python
# только морфологические параметры
def prepareToInsertOnlyMorf(arrayParseSentences, manyMorf):
    for curSentence in arrayParseSentences:
        for curWord in curSentence:
            flagNewInsert = False
            if curWord.dom != '_root':
                mainWord = curSentence[int(curWord.dom) - 1]
                if curWord.feat != 'PR' and mainWord.feat != 'PR':
                    # print(mainWord.text, curWord.text)
                    newMainWord1 = mainWord
                    newDepWord1 = curWord
                    newInserter = OnlyTwoMorf(mainWord.feat, curWord.feat)
                    flagNewInsert = True
                if flagNewInsert:
                    for curInserter in manyMorf:
                        if curInserter == newInserter:
                            curInserter.count += 1
                    else:
                        manyMorf.append(newInserter)
    return manyMorf
```
