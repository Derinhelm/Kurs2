```python
def parseXML(nameFile):
    doc = xml.dom.minidom.parse(nameFile)
    # а зачем храним id(на всякий случай), вообще-то можно и не хранить, id = index + 1
    arrayParseSentences = []
    parent = doc.getElementsByTagName('S')
    for item in parent:
        newSentence = []
        flagNormalWords = True # флаг отсутствия необычных слов(например, пропущенных и тп)
        #слов, у которых нет значения, т.е. элипсис и одно слово в Grechko.tgt
        for child in item.getElementsByTagName('W'):
            if len(child.childNodes) == 0:
                flagNormalWords = False
#<W DOM="5" FEAT="V СОВ ИЗЪЯВ ПРОШ МН" ID="9" LEMMA="ОТВОДИТЬ" LINK="сочин" NODETYPE="FANTOM"/>
            else:
                newNode = Node(child.getAttribute('ID'), child.getAttribute('DOM'), \
                           child.getAttribute('FEAT'), child.childNodes[0].nodeValue, child.getAttribute('LEMMA'))
                newSentence.append(newNode)
        if flagNormalWords:
            arrayParseSentences.append(newSentence)
    return arrayParseSentences
```

```python
class Inserter:
    def __init__(self, mf, mt, mtn, df, dt, dtn):
        self.mainFeat = mf
        self.mainText = mt
        self.mainTextNorm = mtn
        self.depFeat = df
        self.depText = dt
        self.depTextNorm = dtn
        self.count = 1
    def __eq__(self, other):
        return self.mainFeat == other.mainFeat and self.depFeat == other.depFeat and \
            self.mainText == other.mainText and self.depText == other.depText and \
            self.mainTextNorm == other.mainTextNorm and self.depTextNorm == other.depTextNorm
```

```python
def eqNormForm(s1, s2, parseS2):
    #равенство начальных форм слов
    # parseS2 - класс для превращения слова, разобранного в pymorphy2, в несов.вид
    if s1 == s2:
        return True
    if (s1 == s2 + "ся"): # корпус убирает ся, pymorphy2 нет
        return True
    if s1.replace("ё", "е") == s2.replace("ё", "е"):
        return True
    s2Impf = parseS2.inflect({'INFN','impf'})
    if s2Impf != None and parseS2.tag.POS == 'INFN' and s1 == s2Impf.word:
        # приводим второе слово в несов.вид
        print(s1, s2)
        return True
    
    return False
    
    
```

```python
def parseCorporaTag(tagCorpora):
    #возвращает 1. список лямбда-функций, которые надо применить к морфу
    #для проверки на адекватность
    # set важных параметров
    checkFuns = []
    impFeatures = set()
    for curParam in tagCorpora:
        if curParam == 'СРАВ':
            checkFuns.append(lambda m: m.s_cl  == 'comparative')
        elif curParam == 'КР':
            checkFuns.append(lambda m: m.s_cl  == 'shortadjective' or m.s_cl == 'shortparticiple')  
        elif curParam == 'NID':
            return ([lambda m: False], None)
        elif not curParam in ['СЛ', 'COM', 'СМЯГ', 'НЕСТАНД', 'МЕТА', 'НЕПРАВ']:
            impFeatures.add(сorporaToImpMorf[curParam])
    return (checkFuns, impFeatures)
```

```python
def getParseByPymorphy(curWord, curTagCorpora, curNormalForm, arrParse):
    isParsed = False # был ли хоть один нормальный разбор
    (checkFuns, impFeatures) = parseCorporaTag(curTagCorpora)
    notImpFeat = copy.copy(Morf.names) - impFeatures
    goodParsePymorphy = []
    for curParse in arrParse:
        if eqNormForm(curParse.normal_form, curNormalForm, curParse) or \
                'НЕСТАНД' in curTagCorpora or 'НЕПРАВ' in curTagCorpora:
            m = parseToMorf(curWord, curParse)
            if m != None:
                flagTrueParse = True
                for curCheckFun in checkFuns:
                    if not curCheckFun(m):
                        flagTrueParse = False
                        break
                        
                if (flagTrueParse):
                    isParsed = True
                    for curField in notImpFeat:
                        setattr(m, curField, 'not_imp')
                    if not m in goodParsePymorphy:
                        goodParsePymorphy += [m]
        if not isParsed: # pymorphy2 не дал разбора, удовлетворяющего корпусу, пока не рассматриваем это слово
            print("Разборы не соответствуют", curWord, curTagCorpora, curNormalForm)
            return None    
    return goodParsePymorphy
```

```python
def getParseByPymorphyWithAnalize(curWord, curFeat, curNormalForm, arrParse):
    bad = []
    badNormal = []
    goodParsePymorphy = []
    nid = []
    noGoodParse = [] # pymorphy не выдал ни одного разбора, похожего на корпус
    isParsed = False # был ли хоть один нормальный разбор
    for curParse in arrParse:
        if eqNormForm(curParse.normal_form, curNormalForm, curParse) or \
                'НЕСТАНД' in curFeat or 'НЕПРАВ' in curFeat:
            m = parseToMorf(curWord, curParse)
            if m == None:
                bad += [curParse]
            else:
                impFeatures = set()
                flagTrueParse = True 
                # False - если в корпусе сказано, что слово - краткое, а оно нет,
                # или, что сравнительное, а оно нет
                for curParam in curFeat:
                    if curParam == 'СРАВ' and m.s_cl != 'comparative':
                        flagTrueParse = False
                        break
                    elif curParam == 'КР' and (m.s_cl != 'shortadjective' and m.s_cl != 'shortparticiple'):
                        flagTrueParse = False
                        break
                    elif curParam == 'NID':
                        nid += [curWord]
                        break
                    elif not curParam in ['СЛ', 'COM', 'СМЯГ', 'НЕСТАНД', 'МЕТА', 'НЕПРАВ']: 
                        impFeatures.add(сorporaToImpMorf[curParam])
                if (flagTrueParse):
                    isParsed = True
                    notImpFeat = copy.copy(m.names) - impFeatures
                    for curField in notImpFeat:
                        setattr(m, curField, 'not_imp')
            if not m in goodParsePymorphy:
                goodParsePymorphy += [m]
            if not isParsed:
                noGoodParse += [(curWord, curNormalForm, curParse.normal_form, arrParse)]
        else:
            badNormal += [(curWord, curNormalForm, curParse.normal_form)]
    return (goodParsePymorphy, badNormal, bad, nid, noGoodParse)
```

```python
def funWithAnalyze(manyInserter):
    bad = []
    badNormal = []
    goodParsePymorphy = []
    nid = []
    noGoodParse = []
    for i in range(len(manyInserter)):
        curWord = manyInserter[i].mainText
        curFeat = manyInserter[i].mainFeat
        curNormalForm = manyInserter[i].mainTextNorm
        arrParse = morph.parse(curWord)
        g, bn, b, ni, ngp = getParseByPymorphyWithAnalize(curWord, curFeat, curNormalForm, arrParse)
        bad += b
        badNormal += bn
        goodParsePymorphy += g
        nid += ni
        noGoodParse += ngp
    return (bad, badNormal, goodParsePymorphy, nid, noGoodParse)
```
