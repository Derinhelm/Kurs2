```python
f = open("dict.opcorpora.txt")
i = 0
newParadigma = False
words = []
badTag = ['Abbr', 'Name', 'Surn', 'Patr', 'Geox', 'Orgn', 'Trad', 'Erro', 'Dist', 'Infr', \
         'Slng', 'Arch', 'Init', 'Hypo']
for line in f:
        if (newParadigma == True):
                newWord = line.split()[0]
                if (not newWord in words):
                    countBad = 0
                    for bt in badTag:
                        countBad += line.count(bt)
                    #print(line[:-1], countBad)
                    if (countBad == 0):
                        words.append(newWord)
                newParadigma = False
                
        else:
            flagDigit = True
            for j in line:
                if (j == " " or j == "\n"):
                    break
                elif (not (j >= '0' and j <= '9')):
                    flagDigit = False
                    break
            if (flagDigit == True and line != "\n"):
                newParadigma = True
        i += 1
        if (i % 10000 == 0):
            print(i)
        #if (i > 900000):
        #    break
```
